```python
#filling missing valuevwith mean of column
data['mpg_fillNA'] = data['mpg'].fillna(data['mpg'].mean())
```

```python
data.head()
```

```python
# Renaming columns
data = data.rename(columns = {'wt' : 'wgt'})
data.head()
```

```python
# change dtype
data['mpg_new'] = data['mpg_fillNA'].astype(int)
data.head()
```

```python
data['cyl_sqr'] = data['cyl'].apply(lambda x : x*2)
data.head()
```

```python
# Data aggregating and grouping
grouped_mean = data.groupby('mpg')['disp'].mean()
print(grouped_mean)
```

```python
grp_sum = data.groupby(['disp','wgt'])['hp'].sum()
print(grp_sum)
```

```python
# merging and joining dataframe
df1 = pd.DataFrame({'key' : ['A', 'B', 'C'], 'value' : [1, 2, 3]})
df2 = pd.DataFrame({'key' : ['A', 'B', 'D'], 'value' : [4, 5, 6]})
print(df1)
print(df2)
```

```python
#merge df on 'key col'
df = pd.merge(df1, df2, on = "key", how = "inner")
df
```

```python
df = pd.merge(df1, df2, on = "key", how = "outer")
df
```

```python
df = pd.merge(df1, df2, on = "key", how = "left")
df
```

```python
df = pd.merge(df1, df2, on = "key", how = "right")
df
```
